```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
transpot = pd.read_csv('../../processed_data/PPMI/Transcriptomics/transpot_all.csv',).drop("Unnamed: 0", axis=1)
img = pd.read_pickle('../../data/PPMI/MRI_nobrain/mri_meta_2D_4class_nobrain.pkl')[["img_array", "PATNO Visit", "label"]]
```

```python
img
```

```python
img = img.reset_index(drop=True)
```

```python
img
```

```python
img['img_array'][0].shape
```

```python
overlap=pd.merge(transpot,img,on='PATNO Visit',how='inner')
```

```python
overlap
```

```python
overlap=overlap.drop(['label_x'],axis=1)
```

```python
overlap
```

```python
overlap=overlap.rename(columns={'label_y':'label'})
```

```python
overlap
```

```python
overlap.to_csv('../../processed_data/PPMI/overlap/all.csv')
```
